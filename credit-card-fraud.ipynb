```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt


from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## import dataset ##


```python
file = pd.read_csv("/kaggle/input/creditcardfraud/creditcard.csv")
```

```python
file.head(10)
```

```python
file.describe()
```

```python
file.isnull().sum()
```

```python
file['Class'].value_counts()
```

```python
normal = file[file.Class==0]
```

```python
fraud = file[file.Class==1]
```

```python
print(normal.shape)
```

```python
normal.Amount.describe()
```

```python
file.groupby('Class').mean()
```

```python
normal_sample = normal.sample(n = 500)
```

```python
new_file = pd.concat([normal_sample, fraud], axis = 0)
```

```python
new_file.head(10)
```

```python
new_file.groupby('Class').mean()
```

```python
X = new_file.drop(columns = 'Class', axis = 1)
```

```python
Y = new_file['Class'] 
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, stratify = Y, random_state = 2)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
```

```python
training_data_accuracy = accuracy_score(X_train_prediction, Y_train) * 100
```

```python
print(f"Training Data Accuracy: {training_data_accuracy}%")
```
